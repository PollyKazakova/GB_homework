```python
#Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
#Напишите код на Python, реализующий расчет длины вектора, заданного его координатами

import math

a = [10, 10, 10]
b = [0, 0, -10]

len_a = math.sqrt(a[0] * a[0] + a[1] * a[1] + a[2] * a[2])
print(f'Длина вектора a равна: {len_a}')

len_b = math.sqrt(b[0] * b[0] + b[1] * b[1] + b[2] * b[2])
print(f'Длина вектора b равна: {len_b}')
```

```python
#Напишите код на Python, реализующий построение графиков:окружности,эллипса,гиперболы.

from matplotlib.patches import Circle, Ellipse
import matplotlib.pyplot as plt

#Создается область для рисования, связанная с осями координат с помощью метода gca()
ax = plt.gca()

#окружность
#создает круг с центром в точке (x, y), радиуса radius
circle = Circle((6, 7), 5, color = 'blue', fill = False)
ax.add_patch(circle)
plt.axis('scaled')

show()
```

```python
from matplotlib.patches import Circle, Ellipse
import matplotlib.pyplot as plt

#эллипс
ax = plt.gca()

#создает эллипс с центром в точке (x, y),  диаметр которого по горизонтальной оси равен width, а диаметр по вертикальной - height
ellipse = Ellipse((4, 7), 5, 2, color = 'blue', fill = False)
ax.add_patch(ellipse)
plt.axis('scaled')

show()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

#эллипс
t=np.linspace(-5, 5, 100)
X,Y=np.meshgrid(t,t)
plt.figure()

#уравнение эллипса
plt.contour(X, Y, X**2/9+Y**2/4-1, [0], linewidths=3,colors='r')
plt.gcf().set_facecolor('w');
plt.gcf().gca().axis('image');
plt.gcf().gca().grid(True)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

#гипербола
t=np.linspace(-5, 5, 100)
X,Y=np.meshgrid(t,t)
plt.figure()

#уравнение гиперболы
plt.contour(X, Y, X**2/4-Y**2/9-1, [0], linewidths=3,colors='b')
plt.gcf().set_facecolor('w');
plt.gcf().gca().axis('image');
plt.gcf().gca().grid(True)
```

```python
#Нарисуйте трехмерный график двух параллельных плоскостей

from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (8, 8))
ax = plt.axes(projection='3d')
X = np.arange(-15, 15, 2)
Y = np.arange(-15, 15, 2)
X, Y = np.meshgrid(X, Y)

#уравнения параллельных плоскостей
Z = 14*X + 215
Z2 = 14*X + 10

ax.plot_wireframe(X, Y, Z, linestyle='--', colors='green', linewidth=1)
ax.plot_wireframe(X, Y, Z2,colors='red')

show()
```

```python
#Нарисуйте трехмерный график двух любых поверхностей второго порядка
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (8, 8))
ax = plt.axes(projection='3d')
X = np.arange(-15, 15, 2)
Y = np.arange(-15, 15, 2)
X, Y = np.meshgrid(X, Y)

#уравнения плоскостей 2-го порядка
Z = 3*(X**2)-2*X*Y+Y**2
Z2 = X**2/2-Y**2/2

ax.plot_surface(X, Y, Z,  linewidth=1, color='orange')
ax.plot_surface(X, Y, Z2,  linewidth=1, color='red')

show()
```
